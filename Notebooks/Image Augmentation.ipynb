```python
# shift underweight
under_number = 1
for image in os.listdir(under_data_dir):
    img_path = os.path.join(under_data_dir, image)
    img = cv2.imread(img_path)

    def fill(img, h, w):
        img = cv2.resize(img, (h, w), cv2.INTER_CUBIC)
        return img

    def horizontal_shift(img, ratio=0.0):
        if ratio > 1 or ratio < 0:
            return img
        ratio = random.uniform(-ratio, ratio)
        h, w = img.shape[:2]
        to_shift = w*ratio
        if ratio > 0:
            img = img[:, :int(w-to_shift), :]
        if ratio < 0:
            img = img[:, int(-1*to_shift):, :]
        img = fill(img, h, w)
        return img

    def vertical_shift(img, ratio=0.0):
        if ratio > 1 or ratio < 0:
            return img
        ratio = random.uniform(-ratio, ratio)
        h, w = img.shape[:2]
        to_shift = h*ratio
        if ratio > 0:
            img = img[:int(h-to_shift), :, :]
        if ratio < 0:
            img = img[int(-1*to_shift):, :, :]
        img = fill(img, h, w)
        return img

    shifted_hori_img = horizontal_shift(img, 0.1)
    shifted_verti_img = vertical_shift(img, 0.1)
    under_filename = under_data_dir + "shifted_hori" + str(under_number) + '.jpg'
    under_filename_2 = under_data_dir + "shifted_verti" + str(under_number) + '.jpg'
    cv2.imwrite(under_filename, shifted_hori_img)
    cv2.imwrite(under_filename_2, shifted_verti_img)
    under_number = under_number + 1
    
    
# standard weight shift
standard_number = 1
for image in os.listdir(standard_data_dir):
    img_path = os.path.join(standard_data_dir, image)
    img = cv2.imread(img_path)

    def fill(img, h, w):
        img = cv2.resize(img, (h, w), cv2.INTER_CUBIC)
        return img

    def horizontal_shift(img, ratio=0.0):
        if ratio > 1 or ratio < 0:
            return img
        ratio = random.uniform(-ratio, ratio)
        h, w = img.shape[:2]
        to_shift = w*ratio
        if ratio > 0:
            img = img[:, :int(w-to_shift), :]
        if ratio < 0:
            img = img[:, int(-1*to_shift):, :]
        img = fill(img, h, w)
        return img

    def vertical_shift(img, ratio=0.0):
        if ratio > 1 or ratio < 0:
            return img
        ratio = random.uniform(-ratio, ratio)
        h, w = img.shape[:2]
        to_shift = h*ratio
        if ratio > 0:
            img = img[:int(h-to_shift), :, :]
        if ratio < 0:
            img = img[int(-1*to_shift):, :, :]
        img = fill(img, h, w)
        return img

    shifted_hori_img = horizontal_shift(img, 0.1)
    shifted_verti_img = vertical_shift(img, 0.1)
    standard_filename = standard_data_dir + "shifted_hori" + str(standard_number) + '.jpg'
    standard_filename_2 = standard_data_dir + "shifted_verti" + str(standard_number) + '.jpg'
    cv2.imwrite(standard_filename, shifted_hori_img)
    cv2.imwrite(standard_filename_2, shifted_verti_img)
    standard_number = standard_number + 1

    
# overweight shift
over_number = 1
for image in os.listdir(over_data_dir):
    img_path = os.path.join(over_data_dir, image)
    img = cv2.imread(img_path)

    def fill(img, h, w):
        img = cv2.resize(img, (h, w), cv2.INTER_CUBIC)
        return img

    def horizontal_shift(img, ratio=0.0):
        if ratio > 1 or ratio < 0:
            return img
        ratio = random.uniform(-ratio, ratio)
        h, w = img.shape[:2]
        to_shift = w*ratio
        if ratio > 0:
            img = img[:, :int(w-to_shift), :]
        if ratio < 0:
            img = img[:, int(-1*to_shift):, :]
        img = fill(img, h, w)
        return img

    def vertical_shift(img, ratio=0.0):
        if ratio > 1 or ratio < 0:
            return img
        ratio = random.uniform(-ratio, ratio)
        h, w = img.shape[:2]
        to_shift = h*ratio
        if ratio > 0:
            img = img[:int(h-to_shift), :, :]
        if ratio < 0:
            img = img[int(-1*to_shift):, :, :]
        img = fill(img, h, w)
        return img

    shifted_hori_img = horizontal_shift(img, 0.1)
    shifted_verti_img = vertical_shift(img, 0.1)
    over_filename = over_data_dir + "shifted_hori" + str(over_number) + '.jpg'
    over_filename_2 = over_data_dir + "shifted_verti" + str(over_number) + '.jpg'
    cv2.imwrite(over_filename, shifted_hori_img)
    cv2.imwrite(over_filename_2, shifted_verti_img)
    over_number = over_number + 1
```
